## Research Question

Is there an association between increased ridership at San Diego Trolley stations and nearby crime rates over time? Are these effects different across neighborhood socioeconomic contexts?

## Background and Prior Work

Many people argue against the implementation of public transportation in their neighborhoods, citing its tendency to bring crime into their communities. But is this really true? Much research has been done looking into the relationship between public transportation and crime rates and different conclusions have been drawn. Some claim that public transportation can “import” crime into wealthy neighborhoods, leading to higher than normal crime rates. Others see this crime “import” effect as unconvincing, and instead find public transportation alleviates socioeconomic stressors which provoke people into doing crime in the first place.

An issue presented when studying public transportation, particularly discerning its impact on crime rates, is the seeming lack of A/B testing. A place that has access to public transportation and one that doesn’t cannot typically be tested against each other without accounting for a multitude of external factors. This makes sense because many things contribute to whether or not a place is given the opportunity to have public transportation or not. In a clever attempt to navigate this issue, David C. Phillips and Danielle Sandler<a name="cite_note-1"></a> [^](#cite_ref-1) wrote a paper analyzing crime rates in the Washington D.C. area when certain transit stations were closed. The paper found that crime decreases across the railway network when stations are closed and that the decrease in crime was felt more in areas where fewer offenders lived. In essence, the closure of stations can inhibit how easily criminals can navigate the network.

Another project, completed by Greg Ridgeway and John MacDonald, looked more holistically at the railway network in Los Angeles over a 27 year span from 1988-2014.<a name="cite_note-2"></a> [^](#cite_ref-2) They conclude that there is little evidence that access to public railways lead to an increase in crime in neighborhoods. While they do concede that criminals could use railways to travel to lower crime, wealthy areas to find victims, they argue that access to public transportation could lower overall crime rates in a metropolitan area. They claim that giving public transit access to people who hail from low socioeconomic areas can help them seek better opportunities rather than resort to crime.

Looking through the lens of public safety, Tanya Lewis writes an article detailing how public transportation results in far fewer deaths and fewer crimes than automobile travel. <a name="cite_note-3"></a> [^](#cite_ref-3) Per 1 billion passenger miles traveled in 2022, automobiles caused 16.8 fatalities in rural areas and 11.5 in urban areas. To contrast, rail transit fatalities were 5.9 per 1 billion miles and many were non-occupant fatalities that resulted from suicide. Additionally, crimes like assault, robbery, and homicide are down on public transportation in 2023 compared to in 2019 while motor vehicle theft and road rage incidents involving guns are up. The article suggests that many people feel more unsafe on public transportation than they should according to the statistics.

Diving deeper into this phenomenon, Anastasia Loukaitou-Sideris and Camille Fink address the fear women feel when using public transportation.<a name="cite_note-4"></a> [^](#cite_ref-4) Women will often avoid certain railways, only use them in the day, or forego public transportation altogether due to fear of victimization. According to Loukaitou-Sideris and Fink, much of this blame should be put into the lack of input women get when designing safety procedures for public transportation. Citing the lack of security, cameras, panic buttons, or other bystanders who could help, the authors argue that women will continue to avoid public transportation so long as there is no greater attempt to prevent crime around transportation infrastructure like bus stops, trains etc. If public transportation does in fact have a way of bringing crime to certain areas, increased security around transit stops could potentially help to mitigate this.

## Hypothesis


We hypothesize that the relationship between San Diego Trolley ridership and nearby crime rates over time will be weak or mixed, rather than uniformly positive or negative. This is because we expect that any observed differences in crime patterns will be driven largely by preexisting neighborhood socioeconomic conditions, rather than the transit use itself. This aligns with prior research suggesting public transit is often blamed for crime patterns that reflect broader structural inequities rather than transit use.

Null Hypothesis: Changes in San Diego Trolley ridership have no association with nearby crime rates over time.

Alternative Hypothesis: Any association between trolley ridership and crime varies across neighborhoods and is more strongly related to socioeconomic context than to ridership levels alone.

## Data

Our ideal dataset would be a relational database containing different entities on transit data for the San Diego Metropolitan Transit System (MTS) Trolley lines, crime reports in San Diego County, and socioeconomic surveys or census data for San Diegan neighborhoods, specifically along the trolley lines.

This research question requires temporal data on transit operations and ridership to relate crime activity trends. Hence, we would want our “transit data” to contain daily data of service supply—the number of trains in operation, the hours they run, any schedule adjustments made, also security assignments to each station, and station-level ridership activity.

The ridership activity goes hand in hand with the “socioeconomic and census data” listed above. To analyze the effects of the San Diego Trolley lines on crime across multiple locations, we would incorporate socioeconomic variables, like median household income, unemployment, or index of accessibility to radial highways or streets. Additionally, we consider rider information, including trip frequency, daily commute time, and transit dependency. This data may further our research to contextualize any relationship between crime rate and public transit by trolley. For example, we could explore trends in ridership demographics, allowing us to investigate if recent line expansions facilitate the movement of crime or if the infrastructure itself is a hotspot for criminal activity.

For each station, we would need crime reports within a reasonable radius of each trolley station. We would especially need the time and location of the incident to estimate if crime may be related to the trolley system in any way. Ideally, these combinations of datasets—transit data, crime reports, census data—would provide sufficient information as to why the crime occurred near the station. 

<br>

Our ideal dataset would have day-to-day granularity. We consider the number of MTS Trolley stations (62), the timeframe (3-5 years or 1096-1827 days), and the crime buffer to capture our necessary information. Ultimately, this database would consist of approximately 113,000 daily observations. This granularity would distinguish the impact of ridership crowding from typical crime rates and provide an adequate sample for testing causal interactions between transit flow and localized criminal activity.

<br>

This dataset of daily transit data, yearly census data, and daily crime reports would be collected by MTS, the City or County of San Diego, the San Diego Association of Governments, and the San Diego Police Department (SDPD). The transit data is collected via contactless payment systems (PRONTO) and service supply from GTFS (General Transit Feed Specification) feeds. The SDPD records all crime activity in its Automated Regional Justice Information System, accessed via the SANCGIS portal or Open San Diego data feeds. These data would be stored in a relational database, where transit data is identified by its stop ID, census data by neighborhood, and crime reports by incident ID or geospatial information, such as timestamp and street location. This data would be spatially joined using Geographic Information System (GIS) methods since we are interested in the location of crime relative to trolley stops.

<br>

The biggest difference between this ideal dataset and actual data would be the temporal and socioeconomic resolution. Ideally, the dataset would contain daily, synchronized information on the immediate demographic profile of riders, transit schedules, and nearby crime rates. Instead, we have to use annual census and equity data as a proxy and stratify this information against the daily trolley and crime activities.

We have several real datasets to consider:

- https://data.sandiego.gov/datasets/transit-stops/ contains basic information on transit stops and stations in San Diego County. This will let us determine the distance radius from each trolley station for any crime incidents. We can use any of the geospatial variables such as `stop_uid` or `stop_id` to uniquely identify trolley stops and `stop_name` or `(stop_lon, stop_lat)` to filter crime incidents near these locations.

- https://www.sdmts.com/business-center/app-developers/real-time-data links to real time data for transit operations in San Diego. Accessing this data requires an API key, which can be requested by submitting personal information and reasons for access. The high-resolution variables such as `occupancy_status, timestamps, delays, arrivals, departures`, etc. are necessary for inferential analysis, allowing us to explore the impact of service-related crowding and standard criminal activity.

- https://www.sandiego.gov/police/data-transparency/crime-statistics/annual-crime-reports provides links to annual crimes by incident and by neighborhood and historical crime reports for San Diego, whereas https://www.crimemapping.com/map/location/San%20Diego,%20CA,%20USA is an interactive map for locating incidents of crime, arrests, and citations in San Diego County. Using these data together, we can extract exact coordinates or locations and timestamps of arrest and citations. This allows us to perform a spatial join with our trolley station and analyze transit services to these incidents.

- https://data.sandiegocounty.gov/dataset/Equity-Report-Data-Demographics/q9ix-kfws/about_data, https://data.sandiegocounty.gov/dataset/Equity-Report-Data-Geography/p6uw-qxpv/about_data are equity reports in San Diego by demographics and geography, respectively. This data has information on commute times and methods of transportation of San Diegans and most notably `Median Amount of Debt`, `Disproportionality Index`, and methods of transportation `indicators`. These variables are necessary for contextualizing any findings about the relationship between crime rate and trolley services.
